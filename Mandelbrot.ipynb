```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_tier):
    z = c
    for n in range(max_tier):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_tier

def create_mandelbrot(width, height, x_min, x_max, y_min, y_max, max_tier):
    img = np.zeros((height, width))
    for x in range(width):
        for y in range(height):
            real = x_min + (x / width) * (x_max - x_min)
            imag = y_min + (y / height) * (y_max - y_min)
            c = complex(real, imag)
            img[y, x] = mandelbrot(c, max_tier)
            if img[y, x] == max_tier:
              break

        plt.imshow(img, cmap='twilight_shifted')
        plt.axis('off')
        plt.show()


width, height = 800, 600
x_min, x_max = 2.0, 1.0
y_min, y_max = -1.5, 1.5
max_tier = 1000

create_mandelbrot(width, height, x_min, x_max, y_min, y_max, max_tier)
```
